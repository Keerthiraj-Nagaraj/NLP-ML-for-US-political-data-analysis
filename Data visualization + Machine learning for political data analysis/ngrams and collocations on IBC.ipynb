## Importing Libraries

```python
import os, sys, glob, string
import nltk
import time
from nltk.util import ngrams
from operator import itemgetter
from collections import Counter
import pickle
from nltk.corpus import stopwords
from nltk.collocations import *
```

```python
ngram = 1
stopwords = set(stopwords.words('english'))
```

```python
[lib,con,neutral]= pickle.load(open('ibcData.pkl','rb'))
```

```python
liberal = []
for tree in lib:
    liberal.append(tree.get_words())
conservative = []
for tree in con:
    conservative.append(tree.get_words())
neu = []
for tree in neutral:
    neu.append(tree.get_words())
```

```python
function_words = []
fp = open ('function_words.txt','r') 
for line in fp:
    function_words.append(line.rstrip('\r\n').lower())
fp.close
```

## Understanding the collocations of different classes

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
```

### bigram and trigram collocation for liberal sentences

```python
lib_words = []
for line in liberal:
    text = line.translate(str.maketrans('', '', string.punctuation))
    text = text.lower().split()
    text = [w for w in text if not w in stopwords and len(w) > 2]
    text = " ".join(text)
    lib_words.extend(nltk.word_tokenize(text))
```

```python
# change this to read in your data
finder = BigramCollocationFinder.from_words(lib_words)

# only bigrams that appear 20+ times
finder.apply_freq_filter(20) 

# return the 20 n-grams with the highest PMI
finder.nbest(bigram_measures.pmi, 20)  
```

```python
find_3grams = TrigramCollocationFinder.from_words(lib_words)
find_3grams.apply_freq_filter(5)
find_3grams.nbest(trigram_measures.pmi,20)
```
